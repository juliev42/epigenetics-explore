```python
import pandas as pd
import os
import glob
```

```python
ucec_meta = pd.read_csv('data/ucec_metadata.csv',index_col = 0)
```

```python
current_path = os.getcwd()
meth_data = 'data/gdc_ucec_meth_beta/'
```

```python
meth_path = os.path.join(current_path, meth_data)
```

```python
df_list = []
limit = 0
for root, dirs, files in os.walk(meth_path):
    for file in files:
        if file.endswith('level3betas.txt'):
            file_path = os.path.join(root, file)
            
            # Process the content of the text file as needed
            print(f'Read content from {file_path}:')
            df = pd.read_csv(file_path, sep="\t", header=None)
            df.set_index(0)
            df.columns = ['site',file]
            df_list.append(df[file])
            limit += 1
            break
    # if limit == 2: 
    #     break

```

```python
meth_df = pd.concat(df_list, axis=1)

meth_df.head()
```

```python
meth_df.set_index(df['site'], inplace = True)
```

```python
meth_df.head()
```

```python
# Look at data missingness

# horizontal white lines represent present CpG sites, vetical white lines represent present subjects

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(meth_df.isna().values, aspect='auto', cmap=plt.cm.gray, interpolation='nearest')
```
